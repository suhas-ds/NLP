```python
def preprocess_text(text):
    # Ensure text is a string
    if not isinstance(text, str):
        text = str(text)
    
    # Lowercase the text
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords and apply lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    # Reassemble text
    return ' '.join(tokens)

# Apply preprocessing to the "Post title" column
df['Cleaned_Post_Content'] = df['Post Content'].apply(preprocess_text)
df['Cleaned_Post_Title'] = df['Post Title'].apply(preprocess_text)
```

```python
df['Sub Board'] = df['Sub Board'].apply(lambda x: x.replace(' ', ''))
df['Sub Board'] = df['Sub Board'].apply(lambda x: x.replace('"', ''))
df['Sub Board'] = df['Sub Board'].apply(lambda x: x.replace("'", ''))
```

```python
# Sentiment Analysis
# !pip install -q vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()

# Initialize the SentimentIntensityAnalyzer
sentiment = SentimentIntensityAnalyzer()

# Define a function to apply sentiment analysis
def analyze_sentiment(text):
    return sentiment.polarity_scores(text)

# Define a function to get the clean sentiment
def get_clean_sentiment(sentiment_scores):
    # Extract the scores excluding 'compound'
    scores = {k: v for k, v in sentiment_scores.items() if k != 'compound'}
    # Get the category with the highest score
    max_category = max(scores, key=scores.get)
    return max_category

# Apply sentiment analysis to the "Post title" column
df['Sentiment'] = df['Cleaned_Post_Content'].apply(analyze_sentiment)

# Create the clean_sentiment column
df['Clean_Sentiment'] = df['Sentiment'].apply(get_clean_sentiment)

# Optionally, expand sentiment scores into separate columns
df['Negative'] = df['Sentiment'].apply(lambda x: x['neg'])
df['Neutral'] = df['Sentiment'].apply(lambda x: x['neu'])
df['Positive'] = df['Sentiment'].apply(lambda x: x['pos'])
df['Compound'] = df['Sentiment'].apply(lambda x: x['compound'])
```

```python
# Define a function to categorize sentiment
def categorize_sentiment_v(polarity):
    if polarity >= 0.05:
        return 'positive'
    elif polarity <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment categorization
df['senti_v'] = df['Compound'].apply(categorize_sentiment_v)
```

```python
# 1. Common Issues
# Frequent topics (most common words in Post Title and Post Content)
# frequent_words = ' '.join(df['Post Title'].fillna('') + ' ' + df['Post Content'].fillna('')).split()
frequent_words = ' '.join(
    df['Cleaned_Post_Title'].fillna('').astype(str) + ' ' + df['Cleaned_Post_Content'].fillna('').astype(str)
).split()

# frequent_words = ' '.join(df['Cleaned_Post_Content'].fillna('').astype(str)).split()


common_words = Counter(frequent_words).most_common(20)

# Repeated problems (most mentioned keywords like 'driver', 'issue', 'error')
keywords = ['driver', 'issue', 'error', 'broken', 'problem']
repeated_issues = {keyword: df['Post Content'].str.contains(keyword, case=False).sum() for keyword in keywords}
```

```python
# 2. Product Focus
# Most discussed products (by Board Name and Sub Board)
most_discussed_products = df['Sub Board'].value_counts().head(10)

# Software vs Hardware issues
software_keywords = ['software', 'driver', 'update', 'install']
hardware_keywords = ['hardware', 'broken', 'repair', 'battery', 'keyboard']
software_issues = df['Post Content'].str.contains('|'.join(software_keywords), case=False).sum()
hardware_issues = df['Post Content'].str.contains('|'.join(hardware_keywords), case=False).sum()
```

```python
# 3. Time Analysis
df['Post DateTime'] = pd.to_datetime(df['Post Time and Date (CST)'], format='%m/%d/%y %H:%M')
df['Post Hour'] = df['Post DateTime'].dt.hour
post_timing_patterns = df['Post Hour'].value_counts().sort_index()
```

```python
# 4. User Frustration Indicators
frustration_keywords = ['stupid', 'blocked', 'annoying', 'frustrated', 'problem']
frustration_indicators = {word: df['Post Content'].str.contains(word, case=False).sum() for word in frustration_keywords}
```

```python
# 5. Support Response Effectiveness

# Convert 'Moderation Date (CST)' and 'Post DateTime' to datetime
df['Moderation Date (CST)'] = pd.to_datetime(df['Moderation Date (CST)'], format='%d/%m/%y')
df['Post DateTime'] = pd.to_datetime(df['Post DateTime'], format='%d/%m/%y')

# Compute the response time
# Ensure to use .dt.date to work with date parts if necessary
# df['Response Time'] = df['Moderation Date (CST)'] - df['Post DateTime'].dt.date

# If you need the response time in days, you can use .dt.days
df['Response Time Days'] = (df['Post DateTime'] -df['Moderation Date (CST)']).dt.days


# df['Response Time'] = pd.to_datetime(df['Moderation Date (CST)'], format='%m/%d/%y') - df['Post DateTime'].dt.date
# response_times = df['Response Time'].dt.days.describe()

response_times = df['Response Time Days'].describe()
```

```python
# Time-Based Analysis
df['Post Month'] = df['Post Time and Date (CST)'].dt.to_period('M')
monthly_posts = df.groupby('Post Month').size()

print("Monthly Post Frequency:")
print(monthly_posts)
```

```python
# Common Issues Identification
# Keyword Analysis
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Cleaned_Post_Content'])
tfidf_feature_names = tfidf.get_feature_names_out()
tfidf_scores = np.asarray(tfidf_matrix.sum(axis=0)).flatten()
keywords = dict(zip(tfidf_feature_names, tfidf_scores))

# Display top keywords
top_keywords = sorted(keywords.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top Keywords:")
for keyword, score in top_keywords:
    print(f"{keyword}: {score}")
```

```python
# Product Performance Analysis

# Ensure all values in 'Post Title' are strings
df['Post Title'] = df['Post Title'].astype(str)

df['Product Type'] = df['Post Title'].apply(lambda x: 'XPS' if 'XPS' in x else ('Inspiron' if 'Inspiron' in x else 'Other'))
df['Issue Type'] = df['Post Title'].apply(lambda x: 'Hardware' if 'hardware' in x.lower() else ('Software' if 'software' in x.lower() else 'Other'))

product_issue_counts = df.groupby(['Product Type', 'Issue Type']).size().unstack()
print("Issue Counts by Product Type:")
print(product_issue_counts)

```

```python
# Customer Frustration Indicators
frustration_words = ['stupid', 'annoying', 'broken']
df['Frustration Score'] = df['Post Content'].apply(lambda x: sum(word in x.lower() for word in frustration_words))
```
